```python
import pandas as pd 
import os
import re
# 전처리할 데이터 불러오기 
df = pd.read_csv('/content/drive/MyDrive/시사랑 최종데이터.csv')
# dff = pd.read_csv('/content/drive/MyDrive/1-800,2401-4000,4801_5600전처리완료.csv')
```

```python
df = df.drop('Unnamed: 0', axis=1)
```

```python
# dfff = pd.concat([df,dff], axis=0, ignore_index=True)
# dfff.to_csv('/content/drive/MyDrive/1-1000,1101-1800,1901-2100,2401-4000,4801_5600전처리완료.csv', index=False)


# # 1-1000,1101-1800,1901-2100,2401-4000,4801_5600전처리완료
```

```python

# null값이 아닌 데이터들만 가져오기 
df = df[df['content'].notnull()] 
df = df[df['author'].notnull()] 
df = df[df['title'].notnull()] 
# 한글 외에 단어,부호들 제거 
def preprocess_sentence(sent):


  sent = re.sub(r"[^가-힣]+", r" ", sent)
  sent = re.sub(r"\s+", " ", sent)
  return sent 

def replace(sent):
  sent = sent.strip()
  return sent 

# 노이즈 문장 제거 
def remove(text):
    text = re.sub(r'( 이 게시물은).+(이동됨)', '', text)
    return text


# 함수적용
df['content'] = df.content.apply(preprocess_sentence)
df['title'] =  df.title.apply(preprocess_sentence)
df['author'] =  df.author.apply(preprocess_sentence)
df['content'] = df.apply(lambda x : re.sub(f"{x['author']}", ' ', x['content']), axis=1)
df['content'] = df.apply(lambda x : re.sub(f"{x['title']}", ' ', x['content']), axis=1)

df['content'] = df.content.apply(replace)
df['author'] = df.author.apply(replace)
df['title'] = df.title.apply(replace)

df['content'] = df.content.apply(remove)
df['content'] = df['content'].apply(lambda x : x + '</s>')
```
