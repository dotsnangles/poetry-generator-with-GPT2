```python
save_path = '/content/saved'
```

```python
perplexity = datasets.load_metric("perplexity")
input_texts = dataset['content'][:50]
input_texts = [s for s in input_texts if s!='']
results = perplexity.compute(model_id=save_path,
                             input_texts=input_texts)
print(list(results.keys()))
print(round(results["mean_perplexity"], 2))
print(round(results["perplexities"][0], 2))
```

KoGPT-2

```python
model = AutoModelForCausalLM.from_pretrained('skt/kogpt2-base-v2')
tokenizer = AutoTokenizer.from_pretrained("skt/kogpt2-base-v2",
  bos_token='</s>', eos_token='</s>', unk_token='<unk>',
  pad_token='<pad>', mask_token='<mask>')
```

```python
model.save_pretrained('kogpt2-base-v2')
tokenizer.save_pretrained('kogpt2-base-v2')
```

```python
perplexity = datasets.load_metric("perplexity")
input_texts = dataset['content'][:50]
input_texts = [s for s in input_texts if s!='']
results = perplexity.compute(model_id='/content/kogpt2-base-v2',
                             input_texts=input_texts)
print(list(results.keys()))
print(round(results["mean_perplexity"], 2))
print(round(results["perplexities"][0], 2))
```
