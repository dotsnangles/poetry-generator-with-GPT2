```python
save_path = '/content/saved'
```

```python
tokenizer = AutoTokenizer.from_pretrained(save_path)
```

```python
model = AutoModelForCausalLM.from_pretrained(save_path)
```

```python
text = '\uC774\uC6D0\uC7AC' #@param {type:"string"}
max_length = 256 #@param {type:"slider", min:64, max:512, step:8}
repetition_penalty = 2.0 #@param {type:"slider", min:1.5, max:2.5, step:0.1}
```

```python
input_ids = tokenizer.encode(text, return_tensors='pt')
gen_ids = model.generate(input_ids,
                        do_sample=True,
                        max_length=max_length,
                        repetition_penalty=repetition_penalty,
                        pad_token_id=tokenizer.pad_token_id,
                        eos_token_id=tokenizer.bos_token_id,
                        bos_token_id=tokenizer.bos_token_id,
                        use_cache=True)
generated = tokenizer.decode(gen_ids[0])
print(generated)
```
